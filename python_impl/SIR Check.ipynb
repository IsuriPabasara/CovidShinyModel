```python
import math

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)


HOSP_DELAY_TIME = 10
HOSP_RATE = 0.15
HOSP_LOS = 7
ICU_DELAY_TIME = 13
ICU_RATE = 0.02
ICU_LOS = 9
VENT_DELAY_TIME = 13
VENT_RATE = 0.01
VENT_LOS = 10
START_INF = 1

TOTAL_POPULATION = 100000
# S0 = 100000
# I0 = 0
R0 = 2.8
GAMMA = 1 / 14
NUM_DAYS = 365
```

```python
def doubleTime(r0, gamma):
    """gets doubling time based on R0 and gamma """
    return(1 / math.log2(r0 * gamma - gamma + 1))

def getBeta(doubling_time, gamma, s0):
    """gets beta based on doubling time, gamma, and S0"""
    g = 2 ** (1 / doubling_time) - 1
    beta = (g + gamma) / s0  # I think there's a typo in the Penn App but the underlying code looks good.
    return(beta)
```

```python
def sir(s0, i0, r0, beta_vector, gamma, num_days,
        hosp_delay_time=HOSP_DELAY_TIME, hosp_rate=HOSP_RATE, hosp_los=HOSP_LOS,
        icu_delay_time=ICU_DELAY_TIME, icu_rate=ICU_RATE, icu_los=ICU_LOS,
        vent_delay_time=VENT_DELAY_TIME, vent_rate=VENT_RATE, vent_los=VENT_LOS):
    """
    Runs SIR model simulation, and then calculates hospitalizations, icu numbers, and ventilator numbers 
    
    Keyword Args:
      s0: Initial Value for Number of Susceptible
      i0: Initial Value for Number of Infected
      r0: Initial Value for Number of Recovered
      beta_vector: Effective Contact Rate
      
    """
    s = np.repeat(np.nan, num_days)
    i = np.repeat(np.nan, num_days)
    r = np.repeat(np.nan, num_days)
    
    s[0] = s0
    i[0] = i0
    r[0] = r0
        
    # Run SIR Model
    for tt in range(num_days - 1):
        """
        From https://penn-chime.phl.io/, click on more Info about the tool
        """
        beta = beta_vector[tt]
        s[tt + 1] = (-1 * beta * s[tt] * i[tt]) + s[tt]  # This is the derivative. Can read more about SIR model if you want to know more
        i[tt + 1] = (beta * s[tt] * i[tt]) - (gamma * i[tt]) + i[tt]
        r[tt + 1] = (gamma * i[tt])  + r[tt]
    
    # This is because beta_vector is the effective contact rate
    # So if i people are infected, and beta_vector is the effective contact rate
    new_infections = (beta_vector * i) * s
    
    # initialize vectors 
    hosp = np.repeat(0.0, num_days)
    icu = np.repeat(0.0, num_days)
    vent = np.repeat(0.0, num_days)
    admit_hosp = np.repeat(0.0, num_days)
    admit_icu= np.repeat(0.0, num_days)
    admit_vent = np.repeat(0.0, num_days)
    discharge_hosp = np.repeat(0.0, num_days)
    discharge_icu = np.repeat(0.0, num_days)
    discharge_vent = np.repeat(0.0, num_days)
    
    # iteratively creates hospitalization/icu/ventilation admission numbers based 
    # on hospital delay time and hospitalization rates
    for tt in range(num_days):
        if tt + hosp_delay_time < len(admit_hosp):
            admit_hosp[tt + hosp_delay_time] = new_infections[tt] * hosp_rate
        if tt + icu_delay_time < len(admit_icu):
            admit_icu[tt + icu_delay_time] = new_infections[tt] * icu_rate
        if tt + vent_delay_time < len(admit_vent):
            admit_vent[tt + vent_delay_time] = new_infections[tt] * vent_rate
            
    # iteratively creates hospitalization/icu/ventilation discharge numbers based on 
    # admission numbers and length of stays
    for tt in range(num_days):
        if tt + hosp_los < len(discharge_hosp):
            discharge_hosp[tt + hosp_los] = admit_hosp[tt]
        if tt + icu_los < len(discharge_icu):
            discharge_icu[tt + icu_los] = admit_icu[tt]
        if tt + vent_los < len(discharge_vent):
            discharge_vent[tt + vent_los] = admit_vent[tt]
  
    # iteratively creates hospitalization/icu/ventilation numbers based 
    # previous day numbers plus admits minus discharges 
    for tt in range(1, num_days):
        hosp[tt] = hosp[tt - 1] + admit_hosp[tt] - discharge_hosp[tt]
        icu[tt] = icu[tt - 1] + admit_icu[tt] - discharge_icu[tt]
        vent[tt] = vent[tt - 1] + admit_vent[tt] - discharge_vent[tt]
        
    dt = pd.DataFrame({
        'day': list(range(num_days)),
        's': s,
        'i': i,
        'r': r,
        'hosp': hosp,
        'admit_hosp': admit_hosp,
        'discharge_hosp': discharge_hosp,
        'icu': icu,
        'admit_icu': admit_icu,
        'discharge_icu': discharge_icu,
        'vent': vent,
        'admit_event': admit_vent,
        'discharge_vent': discharge_vent,
        'new_infections': new_infections,
    })
     
    return(dt)
```

```python
def find_curr_estimates(s0, beta_vector, gamma, num_days, hospitalized,
                        start_inf=START_INF, hosp_delay_time=HOSP_DELAY_TIME, hosp_rate=HOSP_RATE,
                        hosp_los=HOSP_LOS, icu_delay_time=ICU_DELAY_TIME, icu_rate=ICU_RATE, icu_los=ICU_LOS,
                        vent_delay_time=VENT_DELAY_TIME, vent_rate=VENT_RATE, vent_los=VENT_LOS):
    """Finds current estimates of the number of active infections, number recovered, and number """  
    # Starting number of susceptible people
    start_susc = s0 - start_inf
    start_res = 0 
  
    sir_df = sir(start_susc, start_inf, start_res, beta_vector, gamma, num_days, 
                 hosp_delay_time, hosp_rate, hosp_los,
                 icu_delay_time, icu_rate, icu_los,
                 vent_delay_time, vent_rate, vent_los)
  
    # find the difference between hospitalized column and the currently hospitalized number
    sir_df['diff_proj'] = abs(sir_df.hosp - hospitalized)
  
    # hacky 
    # the # hospitalized will be achieved twice according to model 
    # first as the hospitalizations go up, and second as the hospitalizations go down 
    # we want to find the day
    hosp_numbers = sir_df.hosp
    hosp_change = hosp_numbers.diff()
    hosp_change[0] = 0
    sir_df['hosp_change'] = hosp_change
    
    # only looks at when hospitalizations go up
    curr_day_df = sir_df.loc[sir_df.hosp_change > 0,]
    
    # finds the minimum difference between projected and current hospitalizations
    curr_day_df = curr_day_df.loc[curr_day_df.diff_proj == curr_day_df.diff_proj.min(),] 
    
    curr_day = curr_day_df.day.astype(int)
    infection_estimate = curr_day_df.i.astype(int)
    susceptible_estimate = curr_day_df.s.astype(int)
    recovered_estimate = curr_day_df.r.astype(int)
    
    curr_day_dict = {
        'curr_day': curr_day,
        'infection_estimate': infection_estimate, 
        'susceptible_estimate': susceptible_estimate, 
        'recovered_estimate': recovered_estimate
    }
  
    return(curr_day_dict)
```
